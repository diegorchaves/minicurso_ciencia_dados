
## 2. Dataset Fictício

A seguir temos cerca de 50 transações fictícias de um supermercado.
Cada linha representa uma transação com os itens comprados.


```python
data = [
    ['pão', 'manteiga', 'leite'],
    ['pão', 'manteiga'],
    ['leite', 'café'],
    ['pão', 'café', 'açúcar'],
    ['arroz', 'feijão', 'óleo'],
    ['arroz', 'feijão'],
    ['café', 'biscoito'],
    ['pão', 'queijo', 'presunto'],
    ['pão', 'queijo'],
    ['manteiga', 'queijo'],
    ['arroz', 'carne', 'feijão'],
    ['carne', 'cerveja'],
    ['arroz', 'feijão', 'leite'],
    ['café', 'leite', 'açúcar'],
    ['pão', 'leite', 'queijo'],
    ['pão', 'presunto', 'queijo'],
    ['biscoito', 'refrigerante'],
    ['refrigerante', 'batata frita'],
    ['batata frita', 'cerveja'],
    ['carne', 'batata frita'],
    ['arroz', 'óleo', 'sal'],
    ['sal', 'açúcar'],
    ['pão', 'manteiga', 'café'],
    ['pão', 'manteiga', 'leite'],
    ['pão', 'queijo', 'café'],
    ['arroz', 'feijão', 'carne'],
    ['cerveja', 'biscoito'],
    ['arroz', 'sal'],
    ['feijão', 'óleo'],
    ['pão', 'açúcar', 'café'],
    ['pão', 'manteiga', 'açúcar'],
    ['pão', 'leite'],
    ['queijo', 'presunto'],
    ['carne', 'arroz'],
    ['café', 'biscoito', 'refrigerante'],
    ['pão', 'leite', 'manteiga'],
    ['arroz', 'feijão', 'óleo'],
    ['pão', 'café'],
    ['batata frita', 'refrigerante'],
    ['pão', 'queijo'],
    ['manteiga', 'queijo'],
    ['pão', 'queijo', 'manteiga'],
    ['pão', 'açúcar'],
    ['pão', 'café', 'manteiga'],
    ['carne', 'cerveja'],
    ['arroz', 'sal', 'óleo'],
    ['feijão', 'carne'],
    ['biscoito', 'refrigerante'],
    ['café', 'açúcar'],
    ['pão', 'manteiga'],
]
df = pd.DataFrame({'transacao': range(1, len(data)+1), 'itens': data})
df.head()
```


## 3. Pré-processamento

Precisamos transformar a lista de itens em um formato de **cesta (one-hot encoding)**.


```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(data).transform(data)
df_cesta = pd.DataFrame(te_ary, columns=te.columns_)
df_cesta.head()
```


## 4. Aplicando o Apriori

Defina valores de **suporte mínimo** (por exemplo, 0.1) e gere os itens frequentes.


```python
frequent_items = apriori(df_cesta, min_support=0.1, use_colnames=True)
frequent_items.sort_values('support', ascending=False).head()
```


## 5. Gerando Regras de Associação

Use uma **confiança mínima** (por exemplo, 0.3) para extrair regras.


```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.3)
rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head(10)
```
